```python
from selenium import webdriver
from bs4 import BeautifulSoup
```

```python
# 드라이브 불러오기
driver = webdriver.Chrome()
driver.implicitly_wait(10)
# 웹사이트 접속 (멜론 장르별(8) 스테디셀러)
driver.get('https://www.melon.com/chart/index.htm')
```

```python
# 웹파이지 정보 받기
html = driver.page_source
```

```python
# BS이용
soup = BeautifulSoup(html, 'html.parser')

#soup.select('태그명')
#soup.select('.class속성값')
#soup.select('#id속성값')
#soup.select('태그명.class속성값')

#태그 구조 이용
#soup.select('상위태그정보 > 하위태그정보')
#soup.select('더상위태그정보 >상위태그정보 > 하위태그정보')
```

```python
song_list = soup.select('tbody > tr') #100곡
len(song_list)
```

```python
# 제목명, 가수명
for song in song_list:
    title = song.select('div.ellipsis.rank01> span > a')[0].text
    singer = song.select('div.ellipsis.rank02> span > a')[0].text
    print
    print("title:",title)
    print("singer:", singer)
```
